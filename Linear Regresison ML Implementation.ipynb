# Preparing the dataset

```python
dataset = pd.DataFrame(boston_df.data, columns= boston_df.feature_names)
dataset
```

```python
dataset['Price'] = boston_df.target
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
### Exploratory  Data Analysis
## correlation
dataset.corr()
```

```python
import seaborn as sns
#sns.pairplot(dataset)
```

```python
plt.scatter(dataset['CRIM'], dataset['Price'])
plt.xlabel("Crime rate")
plt.ylabel("Price")
```

```python
plt.scatter(dataset['RM'], dataset['Price'])
plt.xlabel("RM")
plt.ylabel("Price")
```

```python
sns.regplot(x = "RM", y ="Price",data= dataset )
```

```python
## Independent and dependent features
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
## train test split
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
# standarizing dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
xtrain_scale = scaler.fit_transform(xtrain) 
```

```python
xtest_scale = scaler.transform(xtest)
```

# Model Trainning

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain_scale, ytrain)
```

```python
# print the coefficients and the intercept
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
reg.get_params()
```

```python
## prediction with test data
reg_pred = reg.predict(xtest_scale)
```

```python
## plot a scatter plot for the prediction
plt.scatter(reg_pred, ytest)
```

```python
residuals = ytest-reg_pred
residuals
```

```python
# plot the residuals

sns.displot(residuals, kind = "kde")
```

```python
# scatter plot with respect to prediction and residuals
# uniform distribution
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(ytest, reg_pred))
print(mean_squared_error(ytest, reg_pred))
```

```python
# R square and adjusted R square
# R square
from sklearn.metrics import r2_score
score = r2_score(ytest, reg_pred)
```
